```python
import pandas as pd

indiana_df = pd.read_csv("indiana_real_estate_investment.csv")
ohio_df = pd.read_csv("ohio_real_estate_investment.csv")

indiana_df["state"] = "Indiana"
ohio_df["state"] = "Ohio"

combined = pd.concat([indiana_df, ohio_df])
```

```python
combined.to_csv("investment_combined.csv", index=False)
```
